```python
import numpy as np
import pandas as pd
from NaiveBayes import NaiveBayes
```

```python
data = {
    'Emily': -1,
    'John': 1,
    'Sarah': -1,
    'Michael': 1,
    'Jessica': -1,
    'David': 1,
    'Emma': -1,
    'James': 1,
    'Olivia': -1,
    'Daniel': 1,
    'Andrew': 1,
    'Sophia': -1,
    'Matthew': 1,
    'Ava': -1,
    'Joshua': 1,
    'Charlotte': -1,
    'Ethan': 1,
    'Mia': -1,
}

X = list(data.keys())
y = list(data.values())

print(len(X), len(y))
```

```python
def featurize(name):
    features = np.zeros(26)
    for char in name:
        features[ord(char) - ord('A')] = 1
    return features

X = np.array([featurize(name.upper()) for name in X])
y = np.array(y)
```

```python
model = NaiveBayes()
predictions = model.fit(X, y)

df_train = pd.DataFrame({'Name': data.keys(), 'Label': np.where(y==1,"Boy","Girl"), 'Predictions': np.where(predictions==1,"Boy","Girl")})
print('Accuracy:', round(np.mean(predictions == y),2))
df_train
```

```python
X_test = {
    'Liam': 1,
    'Noah': 1,
    'Oliver': 1,
    'Elijah': 1,
    'Lucas': 1,
    'Mason': 1,
    'Logan': 1,
    'James': 1,
    'Alexander': 1,
    'Benjamin': 1,
}

y_test = np.array(list(X_test.values()))
X_test_copy = np.array([featurize(name.upper()) for name in X_test.keys()])
```

```python
preds = model.predict(X_test_copy)
```

```python
preds_sex = np.where(preds == 1, 'Boy', 'Girl')

df_results = pd.DataFrame({'Name': list(X_test.keys()), 'Predicted': preds_sex, 'Actual': np.where(y_test == 1, 'Boy', 'Girl')})
print('Accuracy:', round(np.mean(preds == y_test),2))
df_results
```
