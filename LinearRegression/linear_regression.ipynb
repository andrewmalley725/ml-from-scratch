```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from LinearRegression import LinearRegression
```

```python
X = np.random.randn(300, 1)
X_const = np.hstack((X, np.ones((X.shape[0], 1))))

w = np.array([-2,3])
y = np.dot(X_const, w) + np.random.normal(0,1,X.shape[0])

plt.scatter(X_const[:,0], y)
plt.show()
```

```python
model = LinearRegression()
w = model.fit(X,y)
print('Final weights:', w)
```

```python
plt.scatter(X_const[:,0], y)
plt.plot(X_const[:,0], np.dot(X_const, w), color='red')
plt.show()
```
