```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from KNN import KNN
```

```python
X = np.random.randn(1000, 2)
X[:500, 0] += 6
y = np.zeros(1000)
y[500:] = 1

data = pd.DataFrame({'X1': X[:,0], 'X2': X[:,1], 'Y': y})

plt.scatter(data.X1, data.X2, c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2)
```

```python
k=5
model = KNN(k=k)

model.fit(Xtr, ytr)

yhat = model.predict(Xte)

results = pd.DataFrame(Xte, columns=['X1', 'X2'])
results['yhat'] = yhat
results['yte'] = yte
```

```python
print(f'KNN accuracy ({k}): ', np.mean(yhat == yte))
results.head()
```
