```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from MLP import MLP
```

```python
# Generate 1D sine curve data with 2D input
n_samples = 100
x = np.linspace(-np.pi, np.pi, n_samples)  # x values from - to 
X = np.zeros((n_samples, 1))  # Shape: (n_samples, 2)
X[:, 0] = x  # First feature is x
#X[:, 1] = 0  # Second feature is a constant (0)
Y = np.sin(x).reshape(-1, 1)  # Target: sin(x), shape: (n_samples, 1)

# Initialize and train the MLP
specs = [1, 10, 10, 1]  # 2 inputs, two hidden layers with 10 neurons, 1 output
model = MLP(specs, learning_rate=0.01, num_iterations=5000)
model.fit(X, Y)

# Make predictions
predictions = model.predict(X)

# Calculate final MSE
final_loss = model.mse(Y, predictions)
print(f"\nFinal Mean Squared Error: {final_loss:.6f}")

# Plot the true sine curve and MLP predictions
plt.figure(figsize=(10, 6))
plt.plot(x, Y, label='True sin(x)', color='blue', linewidth=2)
plt.plot(x, predictions, label='MLP Prediction', color='red', linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('MLP Fit to sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
